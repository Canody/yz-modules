# 测试 fft_imaffine 接口

使用标准图和标准输出，与本接口输出进行对比

```python
from yzaet import *
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

print(fft_imaffine())

def im_rescale(im: np.ndarray):
    im = (im - im.min()) / im.max()
    return im
```

```python
# 导入标准图像
im = sio.loadmat('../data/std_im.mat')['im']


# fft_imrotate
d = sio.loadmat('../out_std/rot_im.mat')
rotdeg = np.squeeze(d['rotdeg'].astype(np.float32))
im_rot = np.squeeze(d['im_rot'])

im_rot_new = fft_imrotate(im, rotdeg)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))
axs[0].imshow(im_rot, cmap='gray')
axs[0].text(10, 10, 'MATLAB fftImAffine', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[0].set_axis_off()

axs[1].imshow(im_rot_new, cmap='gray')
axs[1].text(10, 10, 'Python fft_imrotate', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[1].set_axis_off()

axs[2].imshow(im_rot - im_rot_new, cmap='gray')
max_diff = np.max(np.abs(im_rescale(im_rot) - im_rescale(im_rot_new)))
axs[2].text(10, 10, f'Difference: {max_diff:.0e}', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[2].set_axis_off()

plt.tight_layout()
plt.savefig('fft_imrotate_compare.png', dpi=300, bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
# 导入标准图像
im = sio.loadmat('../data/std_im.mat')['im']


# fft_imbin
d = sio.loadmat('../out_std/bin_im.mat')
bin_scale = np.squeeze(d['bin_scale'].astype(np.float32))
im_bin = np.squeeze(d['im_bin'])

im_bin_new = fft_imbin(im, bin_scale)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))
axs[0].imshow(im_bin, cmap='gray')
axs[0].text(10, 10, 'MATLAB fftImAffine', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[0].set_axis_off()

axs[1].imshow(im_bin_new, cmap='gray')
axs[1].text(10, 10, 'Python fft_imbin', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[1].set_axis_off()

axs[2].imshow(im_bin - im_bin_new, cmap='gray')
max_diff = np.max(np.abs(im_rescale(im_bin) - im_rescale(im_bin_new)))
axs[2].text(10, 10, f'Difference: {max_diff:.0e}', fontsize=12, color='white',
            verticalalignment='top', horizontalalignment='left',
            fontweight='bold', fontfamily='monospace')
axs[2].set_axis_off()

plt.tight_layout()
plt.savefig('fft_imbin_compare.png', dpi=300, bbox_inches="tight", pad_inches=0)
plt.show()
```
